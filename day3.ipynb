# Part1

```python
import string

common_elements = []
for i, row in data.iterrows():
    rucksack = row[0]
    length = len(rucksack)
    half = round(length/2)
    
    first_comp = rucksack[0:half]
    second_comp = rucksack[half:]
    unique = list(set(first_comp).intersection(set(second_comp)))[0]
    common_elements.append(unique)
    
    #print(length, rucksack, first_comp, second_comp,unique)
print(common_elements)
```

```python
alphabet = list(string.ascii_lowercase)
all_items = alphabet+[x.upper() for x in alphabet]
priority ={lett: i+1 for i, lett in enumerate(all_items)}

sum_priorities = 0
for i, item in enumerate(common_elements):
    #print(item, priority[item])
    sum_priorities+= priority[item]
print(sum_priorities)
```

```python
i_prev = 0
badges = []

for i in range(3, len(elfs)+3, 3):                
    #print(elfs[i_prev:i])
    group = elfs[i_prev:i]
    i_prev = i 
    intersection = list(set(group[0])&set(group[1])&set(group[2]))[0]
    badges.append(intersection)
    print("group badge found", intersection)
            

sum_priorities = 0
for i, item in enumerate(badges):
    #print(item, priority[item])
    sum_priorities+= priority[item]
print(sum_priorities)            
```
