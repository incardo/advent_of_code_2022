# Part 1

```python
def update_H_T_position(line, H, T, T_positions):
    direc = line[0]
    steps = int(line[1])

    direct_dict = {"R":(1, +1), "L":(1, -1), "U":(0, 1), "D":(0, -1)}

    for step in range(0, steps):
        #print("starting" , H, T)    
        #print("step", step)
        axis = direct_dict[direc][0]
        sign = direct_dict[direc][1]

        H[axis]+=1*sign
        T = update_T_position(T, H)
        T_positions.append(T.copy())
        #print("Ending: -- H", H,"T",  T, "Positions", T_positions)

    return H, T, T_positions

def update_T_position(T, H):
    dist_x = H[0]-T[0]
    dist_y = H[1]-T[1]
    
    #print("distances", dist_x, dist_y)
    if (abs(dist_x) >= 1 and abs(dist_y) > 1) or (abs(dist_x) > 1 and abs(dist_y) >= 1):
        T[0]+=1*np.sign(dist_x)
        T[1]+=1*np.sign(dist_y)
    elif abs(dist_x) > 1 and abs(dist_y) == 0: # move horiz
        T[0]+=1*np.sign(dist_x)
    elif abs(dist_x) == 0 and abs(dist_y) > 1: # move vertical
        T[1]+=1*np.sign(dist_y)
        
    return T  
```

```python
H = [0, 0]
T = [0, 0]
T_positions = []
for line in instructions:
    #print("line", line)
    H, T, T_positions = update_H_T_position(line, H, T, T_positions)
    
    unique_positions = [list(tupl) for tupl in {tuple(item) for item in T_positions}]

print(len(unique_positions))
```

# Part 2

```python
def update_H_all_position(line,H,T1,T2,T3,T4,T5,T6,T7,T8,T9,T9_positions):
    direc = line[0]
    steps = int(line[1])

    direct_dict = {"R":(1, +1), "L":(1, -1), "U":(0, 1), "D":(0, -1)}

    for step in range(0, steps):
        #print("-----step", step)
        #print("starting...","H",H,"T1",T1,"T2",T2,"T3",T3,"T4",T4,"T5",T5,"T6",T6,"T7",T7,"T8",T8,"T9",T9,"T9 positions",T9_positions)    
        axis = direct_dict[direc][0]
        sign = direct_dict[direc][1]

        H[axis]+=1*sign
        T1 = update_T_position(T1, H)
        T2 = update_T_position(T2, T1)
        T3 = update_T_position(T3, T2)
        T4 = update_T_position(T4, T3)
        T5 = update_T_position(T5, T4)
        T6 = update_T_position(T6, T5)
        T7 = update_T_position(T7, T6)
        T8 = update_T_position(T8, T7)
        T9 = update_T_position(T9, T8)
        
        T9_positions.append(T9.copy())
        #print("Ending...","H",H,"T1",T1,"T2",T2,"T3",T3,"T4",T4,"T5",T5,"T6",T6,"T7",T7,"T8",T8,"T9",T9,"T9 positions",T9_positions)    

    return H,T1,T2,T3,T4,T5,T6,T7,T8,T9,T9_positions

def update_T_position(T, H):
    dist_x = H[0]-T[0]
    dist_y = H[1]-T[1]
    
    #print("distances", dist_x, dist_y)
    if (abs(dist_x) >= 1 and abs(dist_y) > 1) or (abs(dist_x) > 1 and abs(dist_y) >= 1):
        T[0]+=1*np.sign(dist_x)
        T[1]+=1*np.sign(dist_y)
    elif abs(dist_x) > 1 and abs(dist_y) == 0: # move horiz
        T[0]+=1*np.sign(dist_x)
    elif abs(dist_x) == 0 and abs(dist_y) > 1: # move vertical
        T[1]+=1*np.sign(dist_y)
        
    return T  
```

```python
H = [0, 0]
T1 = [0, 0]
T2 = [0, 0]
T3 = [0, 0]
T4 = [0, 0]
T5 = [0, 0]
T6 = [0, 0]
T7 = [0, 0]
T8 = [0, 0]
T9 = [0, 0]
T9_positions = []
for line in instructions:
    #print("******line", line)
    H,T1,T2,T3,T4,T5,T6,T7,T8,T9,T9_positions = update_H_all_position(line, H, T1,T2,T3,T4,T5,T6,T7,T8,T9,T9_positions)
    unique_positions = [list(tupl) for tupl in {tuple(item) for item in T9_positions}]

print(len(unique_positions))
```
