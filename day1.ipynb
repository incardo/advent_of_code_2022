# Part1

```python
current_elf = 0 
max_elf = 0
for line in lines:
    if line.strip().isdigit():
        current_elf+=int(line)
        #print(current_elf)
    else:
        if current_elf>max_elf:
            max_elf = current_elf
            print("max elf raising", max_elf)
        current_elf = 0
        
print("final max_elf", max_elf)    
```

# Part2

```python
current_elf = 0 
max_elf = 0
max_elfs_list = []
for line in lines:
    if line.strip().isdigit():
        current_elf+=int(line)
        #print(current_elf)
    else:
        max_elf = current_elf
        max_elfs_list.append(max_elf)
#         print("max elf raising", max_elf)
        current_elf = 0

max_elfs_list.sort()
print(max_elfs_list[-1],max_elfs_list[-2],max_elfs_list[-3])
print(sum([max_elfs_list[-1],max_elfs_list[-2],max_elfs_list[-3]]))
```
