```python
score_shape={"rock":1, "paper":2, "scissors":3}

def do_you_win(opponent, you):
    #print("op:", opponent, "you:", you)
    if opponent == "paper" and you == "rock":
        return "lose",score_shape[you]+0
    elif opponent == "rock" and you == "paper":
        return "win",score_shape[you]+6
    if opponent == "rock" and you == "scissors":
        return "lose",score_shape[you]+0
    elif opponent == "scissors" and you == "rock":
        return "win",score_shape[you]+6
    if opponent == "scissors" and you == "paper":
        return "lose" ,score_shape[you]+0    
    elif opponent == "paper" and you == "scissors":
        return "win",score_shape[you]+6
    if opponent == you:
        return "draw",score_shape[you]+3
    
# def play_game(input1, input2):
#     if input1 == input2:
#     elif input1 ==
```

```python
do_you_win("rock", "scissors")
```

```python
overall_score = 0
for index, row in data.iterrows():
    #print(do_you_win(row[0], row[1]))
    overall_score += do_you_win(row[0], row[1])[1]
print("final score", overall_score)
```
