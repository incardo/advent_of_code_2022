```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.expand_frame_repr', True)
```

```python
data = pd.read_csv('day4_input.txt', header = None, sep = ",|-", names= ["start1", "end1", "start2", "end2"])
```

# Part1

```python
def find_fully_contained(start1, end1, start2, end2):
    if start1 >= start2 and end1<= end2:
        return True
    elif start2>= start1 and end2<=end1:
        return True
    else:
        return False
```

```python
fully_contained = 0
for i, rows in data.iterrows():
    if find_fully_contained(rows.start1, rows.end1, rows.start2, rows.end2):
        fully_contained+=1 

print(fully_contained)
```

# Part2

```python
def find_overlap(start1, end1, start2, end2):
    section1 = set([item for item in range(start1, end1+1)])
    section2 = set([item for item in range(start2, end2+1)])
    intersection = section1&section2
    
    if len(intersection)>0:
        return True
    else:
        return False
    
```

```python
overlap = 0
for i, rows in data.iterrows():
    if find_overlap(rows.start1, rows.end1, rows.start2, rows.end2):
        overlap+=1 

print(overlap)
```
